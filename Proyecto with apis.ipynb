```python
import yfinance as yf
import pandas as pd
```

```python
# ectraemos por medio del identificador de la empresa en este caso APple
apple=yf.Ticker("AAPL")
```

```python
apple
```

```python
#obtenemos la inforación del Ticker
apple_info=apple.info
apple_info
```

```python
#probamos otener información con llaves
apple_info["country"]
```

```python
#obtenemos el precio según parametros - Fuctua en todo momento 
apple_share_price_data=apple.history(period="max")
```

```python
apple_share_price_data
```

```python
apple_share_price_data.head()
```

```python
#reestablecemos el index
apple_share_price_data.reset_index(inplace=True)
```

```python
#revisamos los dividendos de la compañia
apple.dividends
```
